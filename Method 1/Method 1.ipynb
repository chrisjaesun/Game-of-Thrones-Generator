```python
word_transitions = {}
beginning_words = []
with open('ARYA.txt', 'r') as fin:
    for line in fin:
        line = line.rstrip()
        se = re.search(r'[A-Z][a-z]+[.,?";:!]\s', line)
        if se:
            x = se.group().split(' ')
            if len(x) == 1:
                continue
            beginning_words += [se.group().split(' ')[1]]
        arr_of_words = line.split(' ')
        for idx in range(len(arr_of_words)):
            if arr_of_words[idx] not in word_transitions:
                word_transitions[arr_of_words[idx]] = Counter()
            if idx == len(arr_of_words) - 1:
                continue
            word_transitions[arr_of_words[idx]].update([arr_of_words[idx + 1]])


for key, value in word_transitions.items():
    factor = sum(value.values())
    if factor == 0:
        continue
    else:
        word = word_transitions[key]
        for k, v in value.items():
            word[k] /= factor
word_transitions
```

```python
def starting_state():
    """
    Start with a random permutation.
    """
    return np.random.choice(beginning_words)


def simulate_MC(num_iters=5000):
    curr_state = starting_state()
    samples = [curr_state]
    for _ in range(num_iters):
        prob = np.random.rand()
        prob_dict = word_transitions[curr_state]
        accum = 0.0
        for key, value in prob_dict.items():
            accum += value
            if prob <= accum:
                curr_state = key
                break
        if accum == 0:
            samples[len(samples) - 1] += '.'
            curr_state = starting_state()
            samples.append(curr_state)
        else:
            samples.append(curr_state)
    return samples

x = simulate_MC()
" ".join(x)
```
